```python
df = pd.read_excel("diamonds_new.xlsx")

df
```

```python
#write code to find the probability of a premium diamond
#P(E) = desired outcome over total sample space

#count all cuts
cut_total = df['cut'].count()

print(f"the total count of cuts is: {cut_total}")
```

```python
#now count premiums
premium = df[df['cut']=="Premium"]['cut'].count()

print(f"the total count of Premium is: {premium}")
```

```python
#now we find the probability
prob = premium/cut_total

print(f"answer {prob}")
```

The probability is 0.25%

Conditional  probability = <br>
P(A|B) = p(a and b) / p(B)

Probability of J given premium = Prob (j and premium) / prob of premium 

```python
#bring out the j and premium
j_premium = (df['cut']=="Premium") & (df['color']=="J")
```

```python
#count them now
j_premium_count = df[j_premium]["cut"].count()
j_premium_count
```

```python
#p(a and b) / p(b)
final = j_premium_count/premium

final
```
